```python
X=df.drop(['customerID','Churn'],axis=1)
y=df[['Churn']]
```

# Exploratory Data Analysis

```python
county=y.value_counts()
county
```

```python
g=sns.countplot(data=df,x='Churn',hue='Churn',palette='Set1',legend=True)
for i in g.containers:
    g.bar_label(i)
plt.xlabel('Churn')
plt.ylabel('Count')
plt.title('Churn Distribution')
plt.show()
```

```python
plt.pie(county, labels=county.index, autopct='%1.1f%%', 
        startangle=90, colors=sns.color_palette('Set1'))
plt.title('Churn Distribution (Percentage)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python

colors = {'Yes': 'red', 'No': 'blue'}

for i, predictor in enumerate(df.drop(columns=['customerID','Churn', 'TotalCharges', 'MonthlyCharges', 'tenure'])):
    plt.figure(i, figsize=(6, 4))
    ax = sns.countplot(data=df, x=predictor, hue='Churn', palette=colors)
    plt.title(predictor)
    
    for container in ax.containers:
        ax.bar_label(container)
    
    plt.tight_layout()
    plt.show()
```

# Data Preprocessing 

```python
# sample oversampling
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler()
```

```python
X,y=ros.fit_resample(X,y)
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder()
```

```python
X=oe.fit_transform(X)
```

```python
y=oe.fit_transform(y)
```

```python
y=y.ravel()
```

# Model Building and Evaluation 

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

# DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#clf=DecisionTreeClassifier(criterion='gini',splitter='best',max_depth=25,min_samples_split=5,min_samples_leaf=1,max_features=20,max_leaf_nodes=None,min_impurity_decrease=0.0)
clf=DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

## DecisionTree Accuracy For test

```python
print(classification_report(y_test,y_pred))
```

```python
y_pred=clf.predict(X_train)
```

## DecisionTree Accuracy For train

```python
print(classification_report(y_train,y_pred))
```

# GridSearchCV

```python
param_dict={
    "criterion":["gini", "entropy"],
    "splitter":["random", "best"],
    "min_samples_split":list(range(2,5)),
    "min_samples_leaf":list(range(1,5)),
    "max_features": list(range(5,15))
}
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(clf,param_grid=param_dict,cv=10,n_jobs=-1)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

## GridSearchCV Accuracy For test 

```python
print(classification_report(y_test,y_pred))
```

## GridSearchCV Accuracy For train

```python
y_pred=grid.predict(X_train)
```

```python
print(classification_report(y_train,y_pred))
```
